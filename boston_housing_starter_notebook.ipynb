```python
def normalize(df, method):
    # data normalization
    df_copy = df.copy()
    headers = df_copy.columns
    
    # with mu and std
    if method.lower() == 'z_score':
        for c in headers:
            mu = df_copy[c].mean(axis=0)
            std = df_copy[c].std(axis=0)
            df_copy.replace(df_copy[c], value=(df_copy[c]-mu)/std, inplace=True)

            return df_copy 
    
    # with max value in each column
    elif method.lower() == 'max':
        for c in headers:
            max_val = df[c].max()
            df_copy.replace(df_copy[c], value=df_copy[c]/max_val, inplace=True)

            return df_copy
    
    # with min-max
    elif method.lower() == 'minmax':
        for c in headers:
            min_val = df[c].min()
            max_val = df[c].max()
            df_copy.replace(df_copy[c], value=(df_copy[c]-min_val)/(max_val-min_val), inplace=True)

            return df_copy
```

```python
normalized_df = normalize(df, method='z_score')
normalized_df.head()
```

# Detecting Outliers

```python
# using Z-score
from scipy import stats

z = np.abs(stats.zscore(df))
print('Z-Score')
print(z)

threshold = 3
print('\nAfter applying threshold')
result = np.where(z > threshold)
print(result)
```

The result will be in 2 arrays: row and column numbers, e.g. <code>z[row][col]</code>
<br>
<code>z[55][1] = 0.41978194</code>

```python
print(z[55][1])
```

```python
# IQR - Interquartile Range (Jangkauan Interkuartil)
Q1 = df.quantile(0.25) # 25%
Q3 = df.quantile(0.75) # 75%
IQR = Q3 - Q1

print(IQR)
```

```python
# find outliers
outliers = (df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
print(outliers)
```

`False` means that the data is valid (a.k.a not outlier) whereas `True` indicates that there are exist outliers in the data

# Removing Outliers (If necessary)

```python
# with z-score
df_with_z = df.copy()
df_with_z = df_with_z[(z < 3).all(axis=1)]
```

```python
print('Before removing outliers:', len(df))
print('After removing outliers:', len(df_with_z))
```

```python
# with IQR score
df_with_iqr = df.copy()
df_with_iqr = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print('Before removing outliers:', len(df))
print('After removing outliers:', len(df_with_iqr))
```

```python
# Polynomial Regression
x = df['lower_status_percentage']
y = df['price']
f = np.polyfit(x, y, deg=3) # 3rd degree polynomial
p = np.poly1d(f)
```

```python
def plot_poly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

```python
plot_poly(p, x, y, 'num_of_rooms')
```

# Model Evaluation & Refinement

```python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
```

```python
df_copy = df.copy()

labels = df_copy['price']
features = df_copy.drop(['price'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.1, random_state=101)

print('Number of training samples:', len(x_train))
print('Number of test/validation samples:', len(x_test))
```

```python
x = x_train[['num_of_rooms']]

lm_model = LinearRegression()
lm_model.fit(x, y_train)

print('Training R2:', lm_model.score(x, y_train))
print('Testing R2:', lm_model.score(x_test[['num_of_rooms']], y_test))
```

```python
crossval = cross_val_score(lm_model, x_train[['num_of_rooms']], y_train, cv=4) # 4 folds
```

```python
crossval
```

```python
print('Mean folds score:', crossval.mean())
print('Standard deviation:', crossval.std())
```
